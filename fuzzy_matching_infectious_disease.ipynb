# Replace Headers

I plan to remove punctuation to rename the column names and other weird characters

```python
new_col_list = []
for col in df.columns:
    for p in '!"#$%&\'()*+,-./:;<=>?@[\\]^_`{|}~†':
        col = col.replace(p,"")
    new_col_list.append(col)
print(new_col_list)
```

```python
#Creating a new column
df.columns = new_col_list
```

# Format Data to a Readable Format 

We try to replace by 0 for missing values values, so that the dataset looks in a readable format. It is also logical to replace these with zero for the variables like influenza variables and other variables. We would take it as a "negative" or NILL and hence replace them with 0

```python
for col in df.columns:#  For each column replace the NAs with 0
    df[col].fillna(0,inplace=True)
```

```python
# From the percentile distributions, it looks like the outliers needs to be treated for the following variables:

col_list = ['Gonorrhea Current week','Gonorrhea Previous 52 weeks Max','Gonorrhea Cum 2019',
'Haemophilus influenza invasive disease all ages all serotypes All ages all serotypes Current week',
'Haemophilus influenza invasive disease all ages all serotypes All ages all serotypes Previous 52 weeks Max',
'Haemophilus influenza invasive disease all ages all serotypes All ages all serotypes Cum 2019',
'Haemophilus influenza invasive disease all ages all serotypes All ages all serotypes Cum 2018',
'Haemophilus influenza invasive disease age 5 years Serotype b Cum 2019',
'Haemophilus influenza invasive disease age 5 years Serotype b Cum 2018']

# we will ceil the outliers to 98th percentile

for col in col_list:
    p=np.percentile(df[col],98)
    df[col]=np.where(df[col].values>=p,p,df[col])
```

```python
#Describe the data after corrections
df.describe()
```

Now we have removed the outliers by ceiling at 98th percentile

# Conduct Fuzzy Matching

```python
#Looking at the first few rows to see what to use for fuzzy matching
df.head()
```

Let's consider Location 2 column and do a cross join

```python
location_list = list(df['Location 2']) # We store the Location 2 variable in the dataframe
```

```python
# We are trying to match each location with other location and find the fuzzy score

location_set = set(location_list) # We remove the duplicates in location_list

for loc1 in location_set:
    for loc2 in location_set:
        if loc1!=0 and loc2!=0 and loc1!=loc2: # Don't want to find the matching score if locations are 0 and location1!=location2
            scr = fuzz.partial_ratio(str(loc1),str(loc2))
            # We print only if the fuzzy match score is greater than 50
            if scr>50:
                print(f"The matching score between {loc1} and {loc2} is {scr}")
```

I used the best possible fuzzy matching with this dataset.  Any of the other would not make sense.
